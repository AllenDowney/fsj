```python
import pandas
import thinkstats2
import thinkplot
import numpy as np
```

Load the data and remove columns we don't need.

```python
df = pandas.io.parsers.read_csv('FSJ Reproductive Output & Dispersal_24Jul.txt', sep='\t', skiprows=1, header=None)
df.columns = ['jayid', 'usfw', 'nest', 'sex', 'source', 'dispersal', 'local', 'maxclass', 'breeder', 'mindate', 'maxdate',
              'risetime', 'acqtype', 'yrs_breed', 'yrs_help', 'num_nests', 'num_eggs', 'num_hatch', 'num_fledge', 'num_indep',
              'num_yearling', 'num_breed']
avg_columns = ['num_nests', 'num_eggs', 'num_hatch', 'num_fledge', 'num_indep',
              'num_yearling', 'num_breed']
norm_columns = ['risetime', 'yrs_breed', 'yrs_help', 'num_nests', 'num_eggs', 'num_hatch', 'num_fledge', 'num_indep',
              'num_yearling', 'num_breed']

for col in ['jayid', 'mindate', 'maxdate', 'acqtype', 'usfw', 'nest', 'source', 'local', 'breeder']:
    del df[col]

df[:10]
```

Make a copy, select only the breeders, and divide the outcome variables by yrs_breed to get average annual outcomes.

```python
norm = df.copy()

norm = norm[norm.yrs_breed > 0]

for col in avg_columns:
    norm[col] /= norm['yrs_breed']
norm[:10]
```

Compute mean and std for each column so we can standardize scores.

```python
mean = norm.mean()
mean.loc['num_eggs']
```

```python
std = norm.std()
std.loc['num_eggs']
```

Convert the outcome variables to standard scores (number of standard deviations from the mean).

```python
for col in norm_columns:
    norm[col] = (norm[col] - mean.loc[col]) / std.loc[col]
norm[:10]
```

Compute a score that quantifies the tendency for success early in the pipeline and failure later.

```python
pos = ['num_nests', 'num_eggs', 'num_hatch', 'num_fledge']
neg = ['num_indep', 'num_yearling', 'num_breed']
norm['score'] = norm[pos].sum(axis=1) - norm[neg].sum(axis=1)
norm[:10]
```

Group by dispersal.

```python
cols = ['num_nests', 'num_eggs', 'num_hatch', 'num_fledge', 'num_indep', 'num_yearling', 'num_breed', 'score', 'dispersal']

groups = norm[cols].groupby('dispersal')
for name, group in groups:
    print name, len(group)
```

Compute the mean outcome for each dispersal group.

Consistent with the hypothesis, s_to_w immigrants are more successful early in the pipeline and less successful later, so their score is substantially higher than local.

w_to_w immigrants are more successful across the board, although the difference disappears in the last stages.  Their score is higher than local, but lower than s_to_w.


```python
means = groups.aggregate(np.mean)
means
```

The test statistic, diff_star, is the difference in mean score between s_to_w and w_to_w.

```python
diff_star = (means.loc['s_to_w'] - means.loc['w_to_w']).score
diff_star
```

By permuting the labels, we can estimate the p-value, which is the chance of seeing a difference as big as diff_star by chance.

It happens more than 20 times out of 100, so the p-value is more than 0.20, which is not statistically significant.

```python
diffs = []
for i in range(100):
    shuffle = norm.copy()
    shuffle.dispersal = np.random.permutation(shuffle.dispersal)
    fake_groups = shuffle[['score', 'dispersal']].groupby('dispersal')
    fake_means = fake_groups.aggregate(np.mean)
    diff = (fake_means.loc['w_to_w'] - fake_means.loc['s_to_w']).score
    if abs(diff) >= abs(diff_star):
        diffs.append(diff)
diffs, len(diffs)
```
