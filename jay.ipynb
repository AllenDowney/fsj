```python
import pandas
import numpy as np
df = pandas.io.parsers.read_csv('FSJ Reproductive Output & Dispersal.csv', skiprows=1, header=None)
df.columns = ['pair', 'years', 'help', 'nests', 'clutches', 'eggs', 'clutch_size', 'hatchlings',
              'fledglings', 'indeps', 'yearlings', 'breeders', 'dispersal']
del df['pair']
df.index += 2
df[:10]
```

```python
repr([str(x) for x in range(22)])
```

```python
df.help[df.help < 0] = np.nan
df[df.help.isnull()]
```

```python
df.nests[df.nests==0] = np.nan
df[df.nests.isnull()]
```

```python
df.clutches[(df.clutches==0) & (df.eggs>0)] = np.nan
df[df.clutches.isnull()]
```

```python
mean = df.mean()
mean.loc['breeders']
```

```python
norm = df.copy()
for col in df.columns[1:-1]:
    norm[col] = df[col] - mean.loc[col]
norm[:10]
```

```python
norm['index'] = norm[['eggs', 'fledglings', 'indeps', 'yearlings', 'breeders']].sum(axis=1)
norm[:10]
```

```python
groups = norm[['index', 'dispersal']].groupby('dispersal')
for name, group in groups:
    print name, len(group)
```

```python
means = groups.aggregate(np.mean)
means
```

```python
diff_star = (means.loc['w_to_w'] - means.loc['s_to_w'])[0]
diff_star
```

```python
diffs = []
for i in range(100):
    shuffle = norm.copy()
    shuffle.dispersal = np.random.permutation(shuffle.dispersal)
    groups = shuffle[['index', 'dispersal']].groupby('dispersal')
    means = groups.aggregate(np.mean)
    diff = (means.loc['w_to_w'] - means.loc['s_to_w'])[0]
    if diff >= diff_star:
        print diff
```
